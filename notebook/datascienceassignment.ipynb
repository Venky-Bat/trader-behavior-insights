# Part B — Analysis

This section analyzes how trader performance and behavior change across different
market sentiment regimes using the Fear & Greed Index.

```python
import pandas as pd
import matplotlib.pyplot as plt

trades = pd.read_csv("historical_data.csv")
fear_greed = pd.read_csv("fear_greed_index.csv")

# convert timestamps
trades['Timestamp'] = pd.to_datetime(trades['Timestamp'], unit='ms')
fear_greed['timestamp'] = pd.to_datetime(fear_greed['timestamp'], unit='s')

trades['date'] = trades['Timestamp'].dt.date
fear_greed['date'] = fear_greed['timestamp'].dt.date

combined = pd.merge(
    trades,
    fear_greed[['date','classification']],
    on='date',
    how='left'
)

combined['win'] = combined['Closed PnL'] > 0
```

## 1. Performance Across Market Sentiment

We evaluate whether trader performance differs during Fear vs Greed market
conditions using three metrics:

- Average PnL
- Win rate
- Loss magnitude (drawdown proxy)

```python
#Performance Table
performance = combined.groupby('classification').agg(
    avg_pnl=('Closed PnL','mean'),
    win_rate=('win','mean'),
    trades=('Closed PnL','count')
)

performance
```

```python
#Performance Chart
performance['avg_pnl'].plot(kind='bar')

plt.title("Average PnL by Market Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Average PnL")

plt.show()
```

 ## 2. Trader Behavior Across Sentiment

Next, we examine whether traders change behavior depending on
market sentiment by analyzing:

- Trade size
- Trading frequency
- Directional bias

```python
# Behavior Table
behavior = combined.groupby('classification').agg(
    avg_trade_size=('Size USD','mean'),
    trades=('Closed PnL','count')
)

behavior
```

```python
#Behavior Chart
behavior['avg_trade_size'].plot(kind='bar')

plt.title("Average Trade Size by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Average Trade Size (USD)")

plt.show()
```

## 3. Trader Segmentation

We segment traders based on their trading frequency to evaluate
whether activity level affects performance.

Two groups are defined:

- Frequent traders
- Infrequent traders

```python
trader_stats = combined.groupby('Account').agg(
    total_trades=('Closed PnL','count'),
    total_pnl=('Closed PnL','sum'),
    win_rate=('win','mean')
)

threshold = trader_stats['total_trades'].median()

trader_stats['segment'] = trader_stats['total_trades'].apply(
    lambda x: 'Frequent' if x > threshold else 'Infrequent'
)

segment_summary = trader_stats.groupby('segment').agg(
    avg_pnl=('total_pnl','mean'),
    avg_win_rate=('win_rate','mean'),
    traders=('total_trades','count')
)

segment_summary
```

```python
# segment chart 
segment_summary['avg_pnl'].plot(kind='bar')

plt.title("Average PnL by Trader Segment")
plt.xlabel("Trader Segment")
plt.ylabel("Average Total PnL")

plt.show()
```

## Key Insights

1. Traders achieve the highest profitability during Greed market sentiment,
indicating that bullish market conditions improve trading outcomes.

2. Trade sizes are larger during Fear and Extreme Greed periods,
suggesting increased risk-taking when market sentiment is strong.

3. Frequent traders outperform infrequent traders in terms of total PnL,
indicating that trading experience and activity level contribute to better performance.

## Trader Segment 1: High vs Low Leverage Traders

To analyze risk-taking behavior, traders are segmented based on their average leverage.
High-leverage traders may take larger risks compared to low-leverage traders.

```python
# create leverage proxy
combined['leverage_proxy'] = combined['Size USD'] / combined['Execution Price']

# calculate trader-level stats
leverage_stats = combined.groupby('Account').agg(
    avg_leverage=('leverage_proxy','mean'),
    total_pnl=('Closed PnL','sum'),
    win_rate=('win','mean')
)

# create segment
threshold = leverage_stats['avg_leverage'].median()

leverage_stats['segment'] = leverage_stats['avg_leverage'].apply(
    lambda x: 'High Leverage' if x > threshold else 'Low Leverage'
)

leverage_summary = leverage_stats.groupby('segment').agg(
    avg_pnl=('total_pnl','mean'),
    avg_win_rate=('win_rate','mean'),
    traders=('avg_leverage','count')
)

leverage_summary
```

```python
leverage_summary['avg_pnl'].plot(kind='bar')

plt.title("Average PnL: High vs Low Leverage Traders")
plt.xlabel("Trader Segment")
plt.ylabel("Average PnL")

plt.show()
```

## Trader Segment 2: Frequent vs Infrequent Traders

Traders are segmented based on their trading frequency to determine whether
activity level impacts profitability.

```python
trader_stats = combined.groupby('Account').agg(
    total_trades=('Closed PnL','count'),
    total_pnl=('Closed PnL','sum'),
    win_rate=('win','mean')
)

threshold = trader_stats['total_trades'].median()

trader_stats['segment'] = trader_stats['total_trades'].apply(
    lambda x: 'Frequent' if x > threshold else 'Infrequent'
)

freq_summary = trader_stats.groupby('segment').agg(
    avg_pnl=('total_pnl','mean'),
    avg_win_rate=('win_rate','mean'),
    traders=('total_trades','count')
)

freq_summary
```

```python
freq_summary['avg_pnl'].plot(kind='bar')

plt.title("Average PnL: Frequent vs Infrequent Traders")
plt.xlabel("Trader Segment")
plt.ylabel("Average PnL")

plt.show()
```

## Trader Segment 3: Consistent Winners vs Inconsistent Traders

Traders are categorized based on their win rate.
Consistent winners maintain a higher percentage of profitable trades.

```python
winner_stats = combined.groupby('Account').agg(
    win_rate=('win','mean'),
    total_pnl=('Closed PnL','sum'),
    trades=('Closed PnL','count')
)

winner_stats['segment'] = winner_stats['win_rate'].apply(
    lambda x: 'Consistent Winner' if x > 0.5 else 'Inconsistent'
)

winner_summary = winner_stats.groupby('segment').agg(
    avg_pnl=('total_pnl','mean'),
    avg_trades=('trades','mean'),
    traders=('win_rate','count')
)

winner_summary
```

```python
winner_summary['avg_pnl'].plot(kind='bar')

plt.title("Average PnL: Consistent vs Inconsistent Traders")
plt.xlabel("Trader Segment")
plt.ylabel("Average PnL")

plt.show()
```

2️ Strategy 1 — Sentiment-Based Position Sizing

## Strategy 1: Sentiment-Based Position Sizing

### Observation from Analysis
The analysis showed that:
- Traders achieved the highest average PnL during **Greed sentiment** periods.
- Trade sizes increased during **Fear and Extreme Greed**, indicating higher risk-taking behavior.

### Strategy Rule
Traders should adjust their position sizes depending on the prevailing
market sentiment measured by the Fear & Greed Index.

| Market Sentiment | Recommended Action |
|------------------|-------------------|
| Fear | Reduce leverage and position size |
| Neutral | Maintain normal trading size |
| Greed | Increase position size cautiously |
| Extreme Greed | Take profits or reduce exposure |

### Rationale
This approach helps control risk during fearful markets while allowing
traders to capitalize on bullish market momentum.

```python
def position_size_rule(sentiment, base_size):
    
    if sentiment == "Fear":
        return base_size * 0.7
    
    elif sentiment == "Greed":
        return base_size * 1.2
    
    else:
        return base_size
```

## Strategy 2: Segment-Based Trading Strategy

### Observation from Analysis
The trader segmentation analysis revealed that:

- **Frequent traders generated significantly higher average PnL**
- **Consistent winners maintained higher win rates**
- **High leverage traders exhibited higher risk exposure**

### Strategy Rule

Different trader segments should adopt different trading approaches.

| Trader Segment | Recommended Strategy |
|----------------|---------------------|
| Frequent traders | Maintain or increase trading activity during Greed markets |
| Infrequent traders | Reduce trading frequency during Fear periods |
| High leverage traders | Reduce leverage during volatile sentiment conditions |

### Rationale

Segment-based strategies recognize that traders have different
experience levels and risk profiles. Adapting trading behavior
to these segments can improve overall trading performance.

```python
def trading_frequency_rule(segment, sentiment, base_frequency):

    if segment == "Frequent" and sentiment == "Greed":
        return base_frequency * 1.2
    
    elif segment == "Infrequent" and sentiment == "Fear":
        return base_frequency * 0.7
    
    else:
        return base_frequency
```
